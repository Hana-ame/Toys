```python
import cv2
import numpy as np
from numpy.lib.shape_base import split
from scipy.special import comb
```

```python
arr = np.zeros((240*5,240*5,3),dtype=np.uint8)

pts = np.array([[0,0],[140,0],[0,140],[240,240]])

n = len(pts) - 1
split = 100 

w = np.ones((split+1,n+1))

t = np.linspace(0, 1, split+1)
tprod = np.ones(t.shape)
for i in range(n+1):
    w[:,i] *= tprod
    tprod *= t
    
talt = 1-t
tprod = np.ones(t.shape)
for i in range(n+1):
    w[:,n-i] *= tprod
    tprod *= talt

b = comb(n, np.arange(0,n+1,1,dtype=np.uint32))
for i in range(n+1):
    w[:,i] *= b[i]
print(w)

line = np.dot(w, pts)
# print(line.astype(np.uint32))
arr = cv2.polylines(arr, [line.astype(np.int32)],False, (255,255,255), 0)
# print(arr)
cv2.imshow("",arr)
if cv2.waitKey(0):
    cv2.destroyAllWindows()
```

```python
def genWeights(n,split=100):
    w = np.ones((split+1,n+1))
    t = np.linspace(0, 1, split+1)
    tprod = np.ones(t.shape)
    for i in range(n+1):
        w[:,i] *= tprod
        tprod *= t
    tprod = np.ones(t.shape)
    for i in range(n+1):
        w[:,n-i] *= tprod
        tprod *= 1-t
    b = comb(n, np.arange(0,n+1,1,dtype=np.uint32))
    for i in range(n+1):
        w[:,i] *= b[i]
    return w
```
